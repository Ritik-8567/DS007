```python
plt.figure(figsize=(10, 6))
sns.countplot(x="Income Group", hue="System of trade", data=df)
plt.title("Income Group-wise Distribution of Trade Systems")
plt.xticks(rotation=45)
plt.xlabel("Income Group")
plt.ylabel("Count")
plt.legend(title="System of Trade")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="Region", hue="External debt Reporting status", data=df, palette="Set2")
plt.title("External Debt Reporting Status by Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="External Debt Reporting")
plt.show()
```

```python
income_group_counts = df['Income Group'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(income_group_counts, labels=income_group_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Income Groups')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df["Income Group"].value_counts().plot(kind="bar")
plt.title("Distribution of Income Groups")
plt.xlabel("Income Group")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
grouped = df.groupby(['Country Code', 'Income Group']).size().unstack(fill_value=0)
```

```python
grouped.plot(kind='bar', figsize=(8, 4))
plt.title("Distribution of Income Groups (Lower Middle and Upper Middle Income)")
plt.xlabel("Income Group")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
pivot_table = df.pivot_table(index='Country Code', columns='Income Group', aggfunc='size', fill_value=0)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)
plt.title("Country-Income Group Relationship")
plt.xlabel("Income Group")
plt.ylabel("Country Code")
plt.show()
```

```python
grouped = df.groupby(['External debt Reporting status', 'Income Group']).size().unstack(fill_value=0)
```

```python
ax = grouped.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title("External Debt Reporting Status by Income Group")
plt.xlabel("External Debt Reporting Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Income Group")
plt.show()
```

```python
country_code1 = "AFG"
country_code2 = "ARG"
```

```python
country_data1 = df[df['Country Code'] == country_code1]
country_data2 = df[df['Country Code'] == country_code2]
```

```python
income_group1 = country_data1['Income Group'].values[0]
income_group2 = country_data2['Income Group'].values[0]
debt_reporting_status1 = country_data1['External debt Reporting status'].values[0]
debt_reporting_status2 = country_data2['External debt Reporting status'].values[0]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot for Country 1
ax1.bar(['Income Group', 'Debt Reporting Status'], [income_group1, debt_reporting_status1])
ax1.set_title(f"Comparison for {country_code1}")
ax1.set_ylabel("Value")

# Plot for Country 2
ax2.bar(['Income Group', 'Debt Reporting Status'], [income_group2, debt_reporting_status2])
ax2.set_title(f"Comparison for {country_code2}")
ax2.set_ylabel("Value")
plt.tight_layout()
plt.show()
```

```python
numerical_columns = ['National accounts base year', 'National accounts reference year', 'PPP survey year']

# Convert relevant columns to numeric (if applicable)
df[numerical_columns] = df[numerical_columns].apply(pd.to_numeric, errors='coerce')

# Calculate correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Create a density plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x="National accounts base year", fill=True)
plt.title("Density Plot of National Accounts Base Year")
plt.xlabel("National accounts base year")
plt.ylabel("Density")
plt.show()
```
